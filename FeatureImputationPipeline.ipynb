```python
import pandas as pd
import numpy as np

from sklearn.impute import SimpleImputer
```

```python
diabetes = pd.read_csv('datasets/diabetes_processed.csv')
diabetes.head(10)
```

```python
diabetes.shape
```

```python
d_features = diabetes.drop('Outcome', axis=1)
d_label = diabetes[['Outcome']]

d_features.head()
```

```python
mask = np.random.randint(0,100, size=d_features.shape).astype(np.bool)
mask = np.logical_not(mask)
```

```python
d_features[mask] = np.nan
d_features.sample(15)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.compose import ColumnTransformer

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
x_train, x_test, y_train, y_test = train_test_split(d_features, d_label, test_size=0.2)
```

```python
transformer = ColumnTransformer(
    transformers=[('features', SimpleImputer(strategy='mean'), [0,1,2,3,4,5,6,7])]
)
# [0,1,2,...] we use it because transformer needs to know which column in dataset we need to work on
```

```python
clf = make_pipeline(transformer, DecisionTreeClassifier(max_depth=4))
```

```python
clf.fit(x_train, y_train)
clf.score(x_train, y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred, y_test)
```
